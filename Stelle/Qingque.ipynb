```r
SCORE <- 100
```

```r
if(SCORE > 95){
    grade <- 'A+'
}else if(SCORE > 90){
    grade <- 'A'
}else if(SCORE > 80){
    grade <- 'B'
}else if(SCORE > 70){
    grade <- 'C'
}else if(SCORE > 60){
    grade <- 'D'
}else{
    grade <- 'F'
}
grade
```
